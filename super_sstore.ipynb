```python
plt.figure(figsize = (5,5))
plt.bar(list(top['Category'].value_counts().keys()),list(top['Category'].value_counts()),color = 'g')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Count of sales category wise')
plt.show()
```

```python
# the above barplot shows that we have a high count of office supplies and let's check the profit category wise 
```

```python
profit_by_category = top.groupby('Category')['Profit'].sum()
```

```python
profit_by_category 
```

```python
list(profit_by_category.keys())
```

```python
list(profit_by_category.values)
```

```python
colors = ['green' if values>0 else 'red'
          
for values in profit_by_category.values]
```

```python
colors
```

```python
plt.figure(figsize = (5,5))
plt.bar(list(profit_by_category.keys()),list(profit_by_category.values),color = 'g',edgecolor='red',alpha = 0.7)
plt.title('Profit/Loss Breakdown by Category')
plt.show()

# the top-10 data depicts that high profit margin is coming from the technology category and vice-versa for furniture.

# the 1st column of bar shows the loss and rest 2 column shows profit margins.
```
