```python
#Plotting the feature column on count plot
```

```python
a = data[['sex','smoker','region']]
```

```python
a
```

```python
for col in a:
    plt.figure(figsize=(5,5))
    sns.countplot(x=col,hue = 'sex',data=a)
    plt.title(f'Count Plot for {col}')
    plt.show()
```

```python
#plotting a scatter plot of numerical columns
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x= 'charges',y='bmi',hue = 'sex',data=data)
plt.show()
```

```python
sns.pairplot(data[['age', 'bmi', 'children', 'charges']])
plt.title("Pair Plot: Age, BMI, Children, Charges")
plt.show()
```

```python
# premium charges of smokers and non-smokers as per the age 
```

```python
sns.lineplot(x='age', y='charges', hue='smoker', data=data)
plt.title("Premium Charges vs Age (Smoker vs Non-Smoker)")
plt.show()
```

```python
# the above oberservation reflects that the premium charges for smoker are increasing as they getting old than
# non-smokers. 
```
